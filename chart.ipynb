```python
import matplotlib.pyplot as plt

count = [116, 116, 116, 111, 104]
labels = ['아름답다', '맑다', '밝다', '푸르다','드넓다']
colors = ['#2E8B57', '#4CA975', '#60BD89', '#6AC793' ,'#74D19D']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
explode = [0.10, 0.10, 0.10, 0 , 0]
plt.title('[ 힐링되는 ]',fontsize=17) 
plt.pie(count, labels=labels, autopct='%.1f%%',  startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()

```
