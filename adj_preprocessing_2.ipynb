```python
# pandas 불러오기
import pandas as pd
# noun_dict.csv 파일 불러오기
adj_dict = pd.read_csv('./dict_except_food.csv', encoding='utf-8')

# df_mecab_noun.csv 불러오기
adj_total = pd.read_csv('./not_food_list.csv', encoding='utf-8', index_col=0)
```

```python
import re
```

```python
adj_dict
```

```python
adj_total
```

```python
adj_total['isinTrue']=adj_total['isinTrue'].str.replace(pat=r'[^\w]',repl=r' ',regex=True)
```

```python
adj_total['isinTrue']=adj_total['isinTrue'].map(lambda x : (re.sub(' +', ' ', x)))
```

```python
adj_total
```

```python
adj_list = adj_dict['adj'].tolist()
```

```python
adj_list
```

```python
adj_total['pre']= adj_total['isinTrue'].apply(lambda x: x.split(' '))
adj_total['in'] = ''
adj_total['same'] = adj_total['pre'].apply(lambda x: list(set(x) & set(adj_list)))
```

```python
for i, x in enumerate(adj_total['pre']):
    basket = []
    for y in x :
        if y in adj_dict:
            basket.append(y)
    adj_total['in'].iloc[i] = basket
```

```python
adj_total=adj_total.set_index('keyword')
```

```python
adj_total['same'].to_csv('./except_food_adj.csv',encoding='utf-8')
```
